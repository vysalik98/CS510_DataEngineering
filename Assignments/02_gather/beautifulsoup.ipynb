```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# url = "https://www.hubertiming.com/results/2023WyEasterLong"
url = "https://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, "lxml")
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
# Print out the text
text = soup.get_text()
# print(text)
```

```python
soup.find_all("a")
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
# Print the first 10 rows for sanity check
rows = soup.find_all("tr")
print(rows[:10])
```

```python
for row in rows:
    row_td = row.find_all("td")
print(row_td)
type(row_td)
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all("td")
    str_cells = str(cells)
    clean = re.compile("<.*?>")
    clean2 = re.sub(clean, "", str_cells)
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
df1 = df[0].str.split(",", expand=True)
df1.head(10)
```

```python
df1[0] = df1[0].str.strip("[")
df1.head(10)
```

```python
col_labels = soup.find_all("th")
```

```python
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
df3 = df2[0].str.split(",", expand=True)
df3.head()
```

```python
frames = [df3, df1]

df4 = pd.concat(frames)
df4.head(10)
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
df5.info()
df5.shape
```

```python
df6 = df5.dropna(axis=0, how="any")
df6.info()
df6.shape
```

```python
df7 = df6.drop(df6.index[0])
df7.head()
```

```python
import re
for col in df7.columns:
    new_col = re.sub(r'[\[\]]', '', col).strip()
    df7 = df7.rename(columns={col: new_col})
df7.head()
```

```python

for col in df7.columns:
    if df7[col].dtype == "object":
        df7[col] = df7[col].str.replace(']', '')

df7.head()
# df7['Team'] = df7['Team'].str.strip(']')
# df7.head()
```

```python
time_list = df7["Time"].tolist()
time_mins = []
for i in time_list:
    i_split = i.split(":")
    if len(i_split) == 3:
        h, m, s = i_split
    elif len(i_split) == 2:
        h = 0
        m, s = i_split
    else:
        raise ValueError("Unexpected time format: {}".format(i))
    time_math = (int(h) * 3600 + int(m) * 60 + int(s)) / 60
    time_mins.append(time_math)
    # print(time_mins)
```

```python
df7["Runner_mins"] = time_mins

for col in df7.columns:
    if df7[col].dtype == "object":
        df7[col] = (
            df7[col]
            .str.replace("\r", "")
            .str.replace("\n", "")
            .str.replace("\s+", " ")
            .str.replace("\.+", ".")
        )

df7.head()
```

```python
df7.describe(include=[np.number])
```

```python
from pylab import rcParams

rcParams["figure.figsize"] = 15, 5
df7.boxplot(column="Runner_mins")
plt.grid(True, axis="y")
plt.ylabel("Time")
plt.xticks([1], ["Runners"])
```

```python
x = df7["Runner_mins"]
ax = sns.distplot(
    x,
    hist=True,
    kde=True,
    rug=False,
    color="m",
    bins=25,
    hist_kws={"edgecolor": "black"},
)
plt.show()
```

```python
f_fuko = df7.loc[df7[" Gender"] == " F"]["Runner_mins"]
m_fuko = df7.loc[df7[" Gender"] == " M"]["Runner_mins"]
sns.distplot(
    f_fuko,
    hist=True,
    kde=True,
    rug=False,
    hist_kws={"edgecolor": "black"},
    label="Female",
)
sns.distplot(
    m_fuko,
    hist=False,
    kde=True,
    rug=False,
    hist_kws={"edgecolor": "black"},
    label="Male",
)
plt.legend()
```
