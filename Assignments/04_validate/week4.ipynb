```python
import pandas as pd

df = pd.read_csv("Hwy26Crashes2019_S23.csv")
df
```

```python
# crash_df = df[df['Record Type'] == 1]
# crash_df
crash_df = df[df["Record Type"] == 1].dropna(axis="columns", how="all")
crash_df
```

```python
# vehicle_df = df[df['Record Type'] == 2]
# vehicle_df
vehicle_df = df[df["Record Type"] == 2].dropna(axis="columns", how="all")
vehicle_df
```

```python
# particp_df = df[df['Record Type'] == 3]
# particp_df
particp_df = df[df["Record Type"] == 3].dropna(axis="columns", how="all")
particp_df
```

```python
# Every crash has a unique crash ID
msg = "Existence assertion: Every crash has a unique crash ID"
for index, row in df.iterrows():
    if row["Record Type"] in ["1", "2", "3"]:
        assert df["Crash ID"].nunique == len(df), print(f"{msg} FAILED")
print(f"{msg} SUCCEEDED")
```

```python
# Every crash should have a vehicle ID
msg = "Existence assertion: Every crash should have a vehicle ID"
new_df = set(df["Vehicle ID"])
if df["Record Type"].any == 2:
    assert len(new_df) == len(vehicle_df), print(f"{msg} FAILED")
else:
    print(f"{msg} SUCCEEDED")
```

```python
# Every crash county is in Oregon
msg = "Limit assertion: Every crash county is in Oregon"
if not (df["County Code"] >= 1).all() & (df["County Code"] <= 36).all():
    print(f"{msg} SUCCEEDED")
else:
    print(f"{msg} FAILED")
```

```python
# Every crash occurred during 2019
msg = "Limit assertion: Every crash occurred during 2019"
df_year = df["Crash Year"] == 2019
for year in df_year:
    if not year == "2019":
        print(f"{msg} SUCCEEDED")
        break
    else:
        print(f"{msg} FAILED")
        break
```

```python
# Every crash has a crash date
msg = "Intra-record assertion: Every crash has a crash date"
for index, row in df.iterrows():
    if row["Record Type"] == 1:
        if pd.isna(row["Crash Day"]) or row["Crash Day"] == "":
            raise AssertionError("All Crashes do not have a date")
print(f"{msg} SUCCEEDED")
```

```python
# Every person involved in a crash has an ID
msg = "Intra-record assertion: Every person involved in a crash has an ID"
assert (particp_df["Participant ID"]).all(), print(f"{msg} FAILED")
print(f"{msg} SUCCEEDED")
```

```python
# Each crash with a latitude must have a longitude
msg = "Intra-record assertion: Each crash with a latitude must have a longitude"
for row in range(len(crash_df)):
    if pd.isnull(crash_df.loc[row, "Latitude Degrees"]) and not pd.isnull(
        crash_df.loc[row, "Longitude Degrees"]
    ):
        raise AssertionError("Each crash with a latitude do not have a longitude")
    else:
        print(f"{msg} SUCCEEDED")
        break
```

```python
# Each crash has atleast one vehicle involved
msg = "Inter-record assertion: Each Crash has atleast one vehicle"
assert not (df["Participant Vehicle Seq#"]).all() == 0, print(f"{msg} FAILED")

print(f"{msg} SUCCEEDED")
```

```python
# Each participant listed in the crash data was a part of known crash
msg = "Inter-record assertion: Each participant listed in the crash data was a part of known crash"
list1 = set(df["Participant ID"])
for index, rows in particp_df.iterrows():
    if rows["Participant ID"] in list1:
        print(f"{msg} SUCCEEDED")
        break
    else:
        print(f"{msg} FAILED")
```

```python
# Every vehicle listed in the crash data was part of a known crash
msg = "Inter-record assertion: Every vehicle listed in the crash data was part of a known crash"
list2 = df["Vehicle ID"].unique()
for index, row in vehicle_df.iterrows():
    if row["Vehicle ID"] not in list2:
        raise AssertionError(
            "Not every vehicle listed in the crash data was part of a known crash"
        )
print(f"{msg} SUCCEEDED")
```

```python
# The majority of crashes in the data involve vehicles used to transport students

# vehicle_df.columns.tolist()
msg = "Summary assertion: The majority of crashes in the data involve vehicles used to transport students"

vehicle_type_count = df["Vehicle Type Code"].value_counts().idxmax()
print(f"Most crashes recorded has the vehicle type code: {vehicle_type_count}")
if vehicle_type_count:
    print(f"{msg} SUCCEEDED")
else:
    print(f"{msg} FAILED")
# print(vehicle_type_count)
```

```python
# There were thousands of crashes but not millions
msg = "Summary assertion: There were thousands of crashes but not millions"
new_list = set(df[list])
num_crashes = len(new_list)
if num_crashes > 1000000 and num_crashes < 1000:
    raise AssertionError("Number of crashes are not reported in thousands")
else:
    print(f"{msg} SUCCEEDED")
```

```python
# Every crash happened in Highway 26
msg = "Summary assertion: Every crash happened in Highway 26"
assert not crash_df["Highway Number"].all() == 26, print(f"{msg} FAILED")
print(f"{msg} SUCCEEDED")
```

```python
# Crashes frequently happen during afternoon(middle of the day)
msg = "Statistical distribution assertion: Crashes frequently happen during afternoon(middle of the day)"

peak_hour = df["Crash Hour"].value_counts().idxmax()
print(f"Peak Hour: {peak_hour}")

if peak_hour:
    print(f"{msg} SUCCEEDED")
else:
    raise AssertionError("Crashes are not frequently during afternoon")
```

```python
# Crash Data is evenly distributed throughout the year
msg = "Statistical distribution assertion: Crash Data is evenly distributed throughout the year"

crashes_per_month = crash_df["Crash Month"].value_counts()
print(crashes_per_month)

crashes_range = crashes_per_month.max() - crashes_per_month.min()
threshold = 10
if crashes_range <= threshold:
    print(f"{msg} SUCCEEDED")
else:
    raise AssertionError(f"{msg} FAILED")
```
