```python
import pandas as pd
from datetime import datetime, timedelta
```

**ONE TRIP | ONE BUS**


```python
# readcsv
trimet_df1 = pd.read_csv("onetrip.csv")
trimet_df1
```

```python
# dropcolumns
trimet_df1 = trimet_df1.drop(
    columns=["EVENT_NO_STOP", "GPS_SATELLITES", "GPS_HDOP"], axis=1
)
trimet_df1
```

```python
# use columns
# created a new dataframe
trimet_df2 = pd.read_csv(
    "onetrip.csv",
    usecols=[
        "EVENT_NO_TRIP",
        "OPD_DATE",
        "VEHICLE_ID",
        "METERS",
        "ACT_TIME",
        "GPS_LONGITUDE",
        "GPS_LATITUDE",
    ],
)
trimet_df2
```

```python
# Defining a func to decode the columns
def create_timestamp(row):
    opd_date = datetime.strptime(row["OPD_DATE"], "%d%b%Y:%H:%M:%S")
    act_time = timedelta(seconds=row["ACT_TIME"])
    timestamp = opd_date + act_time
    return pd.Timestamp(timestamp)


# Creating a new column
trimet_df2["TIMESTAMP"] = trimet_df2.apply(create_timestamp, axis=1)
trimet_df2
```

```python
trimet_df2.drop(columns=["OPD_DATE", "ACT_TIME"])
```

```python
# calculating difference using diff()
trimet_df2["dMETERS"] = trimet_df2["METERS"].diff()
trimet_df2["dTIMESTAMP"] = trimet_df2["TIMESTAMP"].diff()
trimet_df2
```

```python
# calculating speed using lambda and creating a new column


def calc_speed(row):
    return row["dMETERS"] / row["dTIMESTAMP"].seconds if row["dTIMESTAMP"] != 0 else 0


trimet_df2["SPEED"] = trimet_df2.apply(calc_speed, axis=1)
trimet_df2
```

```python
trimet_df2 = trimet_df2.drop(columns=["dMETERS", "dTIMESTAMP"])
trimet_df2
```

```python
val = trimet_df2.at[1, "SPEED"]
trimet_df2.at[0, "SPEED"] = val
trimet_df2
```

```python
statistics = trimet_df2.aggregate({"SPEED": ["min", "max", "mean"]})
print(statistics)
print("---------------\nStats:")
print(trimet_df2["SPEED"].describe())
```

```python
trimet_df2.to_csv("small.csv", index=False)
```

**ONE BUS | MULTIPLE TRIPS**


```python
trimet_df3 = pd.read_csv("oneday.csv")
trimet_df3
```

```python
trimet_df3 = trimet_df3.drop(
    columns=["EVENT_NO_STOP", "GPS_SATELLITES", "GPS_HDOP"], axis=1
)
trimet_df3
```

```python
trimet_df4 = pd.read_csv(
    "oneday.csv",
    usecols=[
        "EVENT_NO_TRIP",
        "OPD_DATE",
        "VEHICLE_ID",
        "METERS",
        "ACT_TIME",
        "GPS_LONGITUDE",
        "GPS_LATITUDE",
    ],
)
trimet_array = trimet_df4.EVENT_NO_TRIP.unique()
trimet_array
```

```python
for i in trimet_df4.EVENT_NO_TRIP.unique():
    data_df = trimet_df4[trimet_df4["EVENT_NO_TRIP"] == i]
    print(str(i) + " : " + str(data_df.shape[0]))
```

```python
result_df = pd.DataFrame()

trimet_df4 = trimet_df4[trimet_df4["VEHICLE_ID"] == 4223]


for i in trimet_df4.EVENT_NO_TRIP.unique():
    # Looping through every trip
    data_df = trimet_df4[trimet_df4["EVENT_NO_TRIP"] == i]
    data_df = data_df.sort_values(by="ACT_TIME", ascending=True)  # Sorting by time
    data_df = data_df.assign(
        TIMESTAMP=data_df.apply(create_timestamp, axis=1)
    )  # creating timestamp
    data_df["dMETERS"] = data_df["METERS"] - data_df["METERS"].shift(
        1
    )  # Calculating difference
    data_df["dTIMESTAMP"] = data_df["TIMESTAMP"] - data_df["TIMESTAMP"].shift(1)
    data_df["SPEED"] = data_df.apply(calc_speed, axis=1)  # Calculating speed
    data_df.loc[data_df.index[0], "SPEED"] = data_df.loc[data_df.index[1], "SPEED"]
    result_df = pd.concat(
        [
            result_df,
            data_df[
                [
                    "EVENT_NO_TRIP",
                    "VEHICLE_ID",
                    "METERS",
                    "GPS_LONGITUDE",
                    "GPS_LATITUDE",
                    "TIMESTAMP",
                    "SPEED",
                ]
            ],
        ],
        axis=0,
    )

display(result_df)
```

```python
result_df.to_csv("largedata.csv", index=False)
```

**What was the maximum speed for vehicle #4223 on February 15, 2023?**


```python
max_speed = result_df[result_df["VEHICLE_ID"] == 4223][
    result_df["TIMESTAMP"].dt.date == pd.to_datetime("2023-02-15")
]["SPEED"].max()

print("The maximum speed for vehicle #4223 on February 15, 2023 : " + str(max_speed))
```

**Where and when did this maximum speed occur?**


```python
if not result_df.empty and (result_df["SPEED"] == max_speed).any():
    max_speed_rows = result_df[result_df["SPEED"] == max_speed]
    print(f"Number of instances with max speed: {len(max_speed_rows)}")
    for i, row in max_speed_rows.iterrows():
        location = (row["GPS_LATITUDE"], row["GPS_LONGITUDE"], row["TIMESTAMP"])
        print(f"The maximum speed of {max_speed} was achieved at {location}")
else:
    print("No records found for the given vehicle and date.")
```

**What was the median speed for this vehicle on this day?**


```python
median_speed = result_df[
    (result_df["VEHICLE_ID"] == 4223)
    & (result_df["TIMESTAMP"].dt.date == pd.to_datetime("2023-02-15"))
]["SPEED"].median()
print("The median speed for vehicle #4223 on February 15, 2023 : " + str(median_speed))
```

**MULTIPLE BUSES | MULTIPLE TRIPS**


```python
df = pd.read_csv("allday.csv")
df.drop(columns=["EVENT_NO_STOP", "GPS_SATELLITES", "GPS_HDOP"], inplace=True)
df["TIMESTAMP"] = df.apply(create_timestamp, axis=1)
display(df)
```

```python
df = df.sort_values(["VEHICLE_ID", "EVENT_NO_TRIP", "TIMESTAMP"])
df["dMETERS"] = df.groupby(["EVENT_NO_TRIP"])["METERS"].diff()
df["dTIMESTAMP"] = df.groupby(["EVENT_NO_TRIP"])["TIMESTAMP"].diff()
display(df)
```

```python
df["dTIMESTAMP"] = pd.to_timedelta(df["dTIMESTAMP"])


def calculate_speed(row):
    return round(row["dMETERS"] / row["dTIMESTAMP"].total_seconds(), 2)


df["SPEED"] = df.apply(calculate_speed, axis=1)
display(df)
```

```python
df["SPEED"] = df.groupby(["EVENT_NO_TRIP"])["SPEED"].fillna(method="bfill")
display(df)
```

```python
df.to_csv("multiple.csv", index=False)
```

**What was the maximum speed for any vehicle on February 15, 2023?**


```python
max = df["SPEED"].max()
print(f"The maximum speed for any vehicle on February 15, 2023 : {max} m\s")
```

**Where and when did this maximum speed occur?**


```python
max = df["SPEED"].max()
max_row = df[df["SPEED"] == max]
location = {
    "lat": max_row["GPS_LATITUDE"].values[0],
    "lon": max_row["GPS_LONGITUDE"].values[0],
    "time": pd.Timestamp(max_row["TIMESTAMP"].values[0]).strftime("%Y-%m-%d %H:%M:%S"),
}
display(location)
```

**Which vehicle had the fastest mean speed for any single trip on this day? Which vehicle and which trip achieved this fastest average speed?**


```python
mean_speeds = df.groupby(["VEHICLE_ID", "EVENT_NO_TRIP"])[
    "SPEED"].mean().reset_index()
max_mean_speed = mean_speeds["SPEED"].max()
fastest_trip = mean_speeds[mean_speeds["SPEED"] == max_mean_speed].iloc[0]
vehicle_id = fastest_trip["VEHICLE_ID"]
trip_id = fastest_trip["EVENT_NO_TRIP"]

print(
    f"The vehicle with the fastest mean speed on February 15, 2023 was vehicle #{vehicle_id}, on trip #{trip_id}, with an average speed of {max_mean_speed:.2f} mph"
)
```
